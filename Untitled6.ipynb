## 1. Cinemática e Dinâmica de Rotação

Quando estamos sendo introduzidos a mecânica, analisamos o movimento trasnlacional, ou seja, o movimento do corpo de um ponto a outro e com base nisso podemos determinar a aceleração, velocidade, posição e tempo do corpo em questão. Dito isso, também determinamos a energia cinetica como sendo $\frac{mv^2}{2}$. Pórem, no caso de uma hélice girando em um ponto, ela vai de um ponto a outro? Ela tem energia cinética?   

A reposta é **Sim**, embora ela não esteja em um movimento traslacional, ela ainda está exercendo movimento e variando sua posição de maneira angular. 

Dito isso, nesse casos, é bem mais conveniente analisar o movimento rotacional e determinar suas ações e variáveis  

## 2 Definindo as variavéis de rotação

O que é rotação?

É o movimento circular que um corpo faz em torno de seu eixo.

$x = R\cos(\phi) \quad \text{e} \quad y = R\sin(\phi)$

Através da imagem acima, podemos perceber que em um movimento circular podemos decompor as variaveis nos eixos x e y, e utilizalas para entender o movimento circular. A posição angular é responsável por indicar os pontos de interesse, indica quantos graus houve de deslocamento em relação a origem. E a posição é dada por:

$$\phi - \text{Posição angular}$$ 

Na cinématica, temos variáves que descrevem o tempo,velocidade e espaço, de maneira analoga, a cinematica rotacional tem suas variavéis que se alteram em relação ao eixo de rotação, são elas:

$ S - \text{Comprimento do arco}$

$\omega - \text{Velocidade angular}$


$\alpha - \text{Aceleração angular}$

Velocidade angular, é uma grandeza que mede a rapidez com que é feito um percurso em sentido circular.
Aceleração angular, é a variação da velocidade angular em relação ao tempo. Para fins didáticos, nessa introdução vamos analisar corpos em que a distancia entre as particulas permaneça a mesma, ou seja, corpos rígidos.  

Obesvando o exemplo acima podemos notar que V se altera no decorrer do percuso, **não** seu modulo e **sim** seu sentido e direção. Portanto, podemos decompor $V$ em $V_x$ e $V_y$,pela segunda lei de newton, isso implica que uma força esteja atuando no corpo para que isto ocorra e nesse caso a força que está atuando é a força centripeta, podemos decompor a aceleração em duas, a aceleração tangencial ,que como prórpio nome sugere, é tangente ao movimento do disco e a aceleração centripeta que aponta para o centro.

Sabendo que a velocidade a angular é a taxa de variação do angulo em relação ao tempo e que a aceleração angular é a taxa de variação da velocidade angular em relação ao tempo, temos que:

$$\omega = \frac{d\theta}{dt} \quad \text{e} \quad \alpha = \frac{d\omega}{dt}$$

Obs.: $\phi = \theta$, no caso dessas imagens.

$$\omega_m= \frac{\delta\theta}{dt} \quad \text{e} \quad \alpha_m = \frac{\delta\omega}{dt}$$

$$\omega_m - \text{Velocidade rotacional média}$$

$$\alpha_m - \text{Aceleração rotacional média}$$

Podemos relacionar as variaveis lineares com as angulares através da seguinte relação:

$$S = \theta R$$  
$$V = \omega R$$  
$$a_t = \alpha R$$
$$a_c = \frac{v^2}{R}$$

Sendo:

$$V - \text{Veolicidade linear} $$
$$a - \text{Aceleração tangencial}$$

Portanto, separando os componentes da variáveis temos que:

$$V_x = \frac{dx}{dt} \quad \text{e} \quad V_y = \frac{dy}{dt}$$

$$V_x = \frac{d\theta R\cos(\theta)}{dt} \quad \text{e} \quad V_y = \frac{d\theta R\sin(\theta)}{dt}$$

$$V_x = \omega R\cos(\theta) \quad \text{e} \quad V_y = \omega R\sin(\theta)$$

## 3 Alguns exemplos para essas váriaveis

Seguem abaixo alguns códigos exemplicando a relação entre cada variavel.

Neste primeiro código, vemos a relação entre posição, velocidade e tempo. Considerando um disco de vinil que gira a uma velocidade angular constante de 2 rad/s em um tempo de 10s podemos determinar a posição final e o número de voltas do disco. 

```python
t = np.linspace(0, 10, 100)
omega = 2  
theta = omega * t

number_turn = theta[99]/(2*np.pi)
print(number_turn) # Número de voltas

plt.plot(t, theta)
plt.xlabel('Tempo em segundos')
plt.ylabel('Ângulo em radianos')
plt.title('Variação do ângulo em função do tempo na rotação de um disco de vinil')
plt.grid(True)
```

Agora o mesmo exemplo do anterior, porém com velocidade angular não constante.

```python
t = np.linspace(0, 10, 100)
omega_0 = 2  
alpha = 0.5  

omega = omega_0 + alpha * t  
theta = np.cumsum(omega) * (t[1] - t[0])  

plt.plot(t, theta)
plt.xlabel('Tempo em segundos')
plt.ylabel('Ângulo em radianos')
plt.title('Variação do ângulo em função do tempo na rotação de um disco de vinil')
plt.grid(True)

plt.show()
```

Segue abaixo, alguns exemplos de velocidades angulares arbitrarias em função do tempo, note que essas as formulas são algumas sem um fenômeno especifico envolivdo afim de servirem como exemplo para o gráfico abaixo.

```python
# Velocidades angulares para cada exemplo
omega_1 = np.sin(t)              # Exemplo 1
omega_2 = 0.5 * np.cos(t)        # Exemplo 2
omega_3 = np.exp(-t/5)           # Exemplo 3
omega_4 = 0.8 * np.sin(2*t)      # Exemplo 4

# Plot do gráfico
plt.plot(t, omega_1, label='Exemplo 1')
plt.plot(t, omega_2, label='Exemplo 2')
plt.plot(t, omega_3, label='Exemplo 3')
plt.plot(t, omega_4, label='Exemplo 4')

plt.xlabel('Tempo em segundos')
plt.ylabel('Velocidade angular ')
plt.title('Gráfico da Velocidade Angular em Função do Tempo')
plt.grid(True)
plt.legend()
plt.show()
```

Segue abaixo, um exemplo demostrando a relação entre aceleração angular e o tempo.

```python
theta = 2 * np.pi * t  # Ângulo em função do tempo
omega = theta * t  # Velocidade angular em função do tempo

# Derivada da velocidade angular em relação ao tempo
alpha = np.gradient(omega, t)

#Plotando o gráfico
plt.plot(t, alpha)
plt.xlabel('Tempo em segundos')
plt.ylabel('Aceleração Angular')
plt.title('Gráfico da Aceleração Angular em Função do Tempo')
plt.grid(True)

plt.show()
```

## 4. Torque e momento de inércia

Torque, consiste na ação de girar ou torcer um corpo em torno do seu eixo de rotação por meio da aplicação de uma força. Para que apliquemos um torque sobre um corpo, é preciso que a força aplicada coincida com o eixo de rotação. Em outras palavras, torque seria "a força que faz girar".

Torque é uma grandeza vetorial e é calculada pelo produto vetorial entre o "braço da alavanca" e a força que está sendo aplicada. Podemos calcular o torque e o módulo pelas seguintes expressões:

$$ \vec{t} = \vec{r} \times \vec{F} \quad \text{e} \quad t=rF\sin(\theta)$$

```python
l = 1
omega = 2

def plot_bastao(angulo):
    fig, ax = plt.subplots()
    
    # Coordenadas do ponto final do bastão
    x = [0, l*np.cos(np.radians(angulo))]
    y = [0, l*np.sin(np.radians(angulo))]
    ax.plot(x, y, color='red')
    
    # Vetor de força (torque)
    torque = 0.5  # Valor do torque aplicado 
    x_torque = [l*np.cos(np.radians(angulo)), l*np.cos(np.radians(angulo)) + torque*np.cos(np.radians(90 + angulo))]
    y_torque = [l*np.sin(np.radians(angulo)), l*np.sin(np.radians(angulo)) + torque*np.sin(np.radians(90 + angulo))]
    ax.arrow(x_torque[0], y_torque[0], x_torque[1] - x_torque[0], y_torque[1] - y_torque[0], 
             head_width=0.1, head_length=0.1, fc='blue', ec='blue')
    
    ax.set_xlim(-1.1, 1.1)
    ax.set_ylim(-1.1, 1.1)
    ax.set_aspect('equal')
    plt.show()

girar = wdg.FloatSlider(value=0.0, min=0.0, max=360.0, step=1, description='Ângulo')
wdg.interactive(plot_bastao, angulo=girar)
```

O gráfico interativo mostra o exemplo de uma força sendo apliacada em um bastão cuja o eixo de rotação se encontra na origem das coordenadas x e y.

Na imagem acima, podemos observar 3 particulas girando em torno de um eixo de rotação com as distancias r1,r2 e r3. Agora, vamos determinar a energia cinética associada a essa imagem. 

Sabemos que $E_c = \frac{mv^2}{2}$ é a energia cinética e $V = \omega r$, então:

$$E_c = \frac{m\omega^2 r^2}{2}$$

No caso da imagem acima temos que:

$$E_c = \frac{m_1\omega^2 r_1^2}{2} + \frac{m_2\omega^2 r_2^2}{2}  + \frac{m_3\omega^2 r_3^2}{2}$$

$$E_c = \frac{\omega^2 (m_1r_1^2+m_2r_2^2+m_3r^3)}{2}$$

$$E_c = \frac{\omega^2(\sum_{i=1}^3 m_ir_i^2)}{2}$$

$$E_c = \frac{\omega^2I}{2}$$

$$I = \sum_{i=1}^n m_ir_i^2$$

Sendo esse $I$, o momento de inércia de um corpo, em outras palavras esse $I$ simboliza a tendência de um corpo  maciço de se manter no seu estado natural de movimento. 

O caso acima se trata do momento de inercia discreto. Para o momento de inércia continuo a expressão é dada por:

$$I = \int_{}^{} r^2 \, dm$$
